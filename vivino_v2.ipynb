# 1. Wine URL 수집하기

```python
import os
import pandas as pd
import time
from selenium.webdriver import Chrome, ChromeOptions
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

# Selenium WebDriver 설정
options = ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--force-device-scale-factor=0.5')

VIVINO_URL = "https://www.vivino.com/explore?e=eJzLLbI11rNQy83MszVQy02ssDU1AAG15Erb7CK1ZFvvoHC1AltDtfQ027LEoszUksQctfyiFNuU1OJktfykStuixJLMvPTi-OT80rwSALirGjE%3D"
all_wine_urls = list()
checkpoint_file = os.path.expanduser("./wine_urls_v2.csv")

def load_and_scrape(driver, VIVINO_URL, pause_time=2):
    driver.get(VIVINO_URL)
    time.sleep(3)  # 초기 페이지 로드 대기

    while True:
        try:
            # 와인 카드 찾기
            wine_cards = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//div[contains(@class, 'wineCard__wineCard--2dj2T')]//a"))
            )
            print(f"Found {len(wine_cards)} wine cards.")

            # 각 카드에서 href 출력 (필터링 적용)
            for card in wine_cards:
                urls = card.get_attribute('href')
                if urls and '/w/' in urls:  # '/w/'가 포함된 URL만 출력
                    print(urls)

            # 페이지 끝으로 스크롤
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # 스크롤 후 잠시 대기

            # 다음 페이지 버튼 클릭
            try:
                next_button = WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'MuiButtonBase-root')]"))
                )
                next_button.click()
                print("Clicked next button.")
                time.sleep(pause_time)  # 버튼 클릭 후 대기
            except TimeoutException:
                print("No next button found or not clickable.")
                break
        except TimeoutException:
            print("No wine cards found.")
            break
        except Exception as e:
            print(f"Unexpected error: {e}")
            break

# WebDriver 실행 및 함수 호출
driver = Chrome(options=options)
load_and_scrape(driver, VIVINO_URL, pause_time=2)
driver.quit()
```
